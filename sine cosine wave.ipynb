```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
y_arr=[]
x_arr=[]
```

```python
f=4
a=5
```

```python
inval=np.linspace(0,2,200)
```

```python
for t in inval:
    s=a*math.cos(2*np.pi*f*t)
    b=a*math.sin(2*np.pi*f*t)
    y_arr.append(s)
    x_arr.append(b)
```

```python
print(y_arr)
```

```python
plt.plot(y_arr, color="red")
plt.plot(x_arr, color="green")
plt.legend(['s = cosine','b = sine'],loc='lower left')
plt.xlabel('time')
plt.ylabel('value')
```
